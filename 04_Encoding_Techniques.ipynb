# **1. One Hot Encoding**

Creates a binary column for each category value. No ordinal assumption. Causes dimensionality explosion for high-cardinality features.

Use when

-> Nominal categorical features

-> Low cardinality (< ~15 categories)

-> Linear models, neural nets

-> Color, gender, city (low cardinality)

Avoid when

Features with 100+ unique values — use target encoding or embedding instead.

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Red"]
})
```

```python
df.head()
```

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoded = encoder.fit_transform(df[["Color"]])
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out()
)
```

```python
print(encoded_df)
```

# **2. Label encoding**

Maps each category to an integer. Introduces an artificial ordinal relationship. Best only for ordinal categories or tree models.

Use when

-> Ordinal categories (low/med/high)

-> Decision trees, random forests, XGBoost

-> Target variable in classification

Avoid when

Do NOT use for nominal data with linear models — implies A < B < C which may be false.

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Red"]
})
```

```python
df.head()
```

```python
encoder = LabelEncoder()
```

```python
df["Color_Encoded"] = encoder.fit_transform(df["Color"])
```

```python
print(df)
```

# **3. Target (mean) encoding**

Replaces each category with the mean of the target variable for that group. Powerful but prone to leakage — needs cross-validation.


Use when

-> High-cardinality categoricals

-> ZIP codes, product IDs, user IDs

-> Gradient boosting models

-> Competition / production ML

Avoid when

Must use with careful cross-validation to prevent data leakage.

# **4. Ordinal encoding**

Manually assigns ordered integers reflecting a known rank relationship between categories.

Use when

-> Education level (HS < UG < PG)

-> Satisfaction ratings

-> Frequency (never/sometimes/always)

Avoid when

Only use when a true, meaningful order exists between categories.

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    "Education": ["Primary", "Graduate", "Secondary", "Post Graduate"]
})
```

```python
df.head()
```

```python
df['Education'].unique()
```

```python
encoder = OrdinalEncoder(
    categories=[["Primary", "Secondary", "Graduate", "Post Graduate"]]
)
```

```python
df["Education_Encoded"] = encoder.fit_transform(df[["Education"]])
```

```python
print(df)
```
