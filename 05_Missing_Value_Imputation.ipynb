**1. Mean Imputation**

Replaces missing values with the column mean. Simple and fast but reduces variance and ignores relationships.

Use when

-> Missing completely at random (MCAR)

-> Small % of missing values

-> Baseline imputation

-> Use median for skewed features

Avoid when

Not suitable when data is missing not at random (MNAR) — creates bias.

```python
ethnic['age'].head()
```

```python
ethnic['age'] = ethnic['age'].fillna(ethnic['age'].mean(), inplace=True)
```

```python
ethnic['age'].head()
```

```python
ethnic['age'].isnull().sum()
```

**2. Median Imputation**

Replaces missing values with the column median. Simple and fast but reduces variance and ignores relationships.

Use when

-> Missing completely at random (MCAR)

-> Small % of missing values

-> Baseline imputation

-> Use median for skewed features

Avoid when

Not suitable when data is missing not at random (MNAR) — creates bias.

```python
ethnic['Salaries'].head(-1)
```

```python
ethnic['Salaries'] = ethnic['Salaries'].fillna(ethnic['Salaries'].median(), inplace=True)
```

```python
ethnic['Salaries'].head(-1)
```

```python
ethnic['Salaries'].isnull().sum()
```

**3. Mode Imputation**

Replaces missing values with the column mode. Simple and fast but reduces variance and ignores relationships.

Use when

-> Missing completely at random (MCAR)

-> Small % of missing values

-> Baseline imputation

-> Use median for skewed features

-> Use mode for Categorical features

Avoid when

Not suitable when data is missing not at random (MNAR) — creates bias.

```python
ethnic['Sex'].head()
```

```python
ethnic['Sex'] = ethnic['Sex'].fillna(ethnic['Sex'].mode()[0])
# ethnic['Sex'].fillna(ethnic['Sex].mode()[0], inplace = True) # if you want to add inplace = True
```

**4. KNN Imputation**

Finds k most similar rows and uses their values to fill the gap. More accurate than mean but computationally expensive.

Use when

-> Small-to-medium datasets

-> When relationships between features matter

-> MICE-like iterative imputation

Avoid when

Computationally expensive for large datasets. Needs feature scaling first.

```python
from sklearn.impute import KNNImputer
```

```python
ethnic['Salaries'].head()
```

```python
imputer = KNNImputer(n_neighbors=5)
```

```python
ethnic['Salaries'] = imputer.fit_transform(ethnic[['Salaries']])
```

```python
ethnic['Salaries'].head()
```

**5. Indicator + Impute**

Creates a binary flag column to indicate missingness, then imputes. Preserves the information that a value was missing.

Use when

-> When missingness itself is informative

-> Medical / survey data

-> Combined with any imputation method

Avoid when

Unnecessary if data is MCAR and you are just quickly imputing for a baseline.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.DataFrame({
    "Salary": [40000, np.nan, 50000, np.nan],
    "Loan_Approved": [True, False, True, False]
})
```

```python
imputer = SimpleImputer(
    strategy="mean",
    add_indicator=True
)
```

```python
result = imputer.fit_transform(df[['Salary']])
print(result)
```
