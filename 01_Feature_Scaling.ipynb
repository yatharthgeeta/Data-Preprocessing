**1. MinMax Scaling**

Rescales every feature to a fixed range, usually [0, 1]. Sensitive to outliers since min/max are determined by the data.

Use when

-> Neural networks & deep learning

-> Image pixel normalization

-> KNN & distance-based models

-> When feature range matters (e.g.

-> ReLU inputs)

-> Algorithms needing bounded input

Avoid when

data has significant outliers — they compress all other values into a tiny range.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data.head()
```

```python
scaler = MinMaxScaler()
x_minmax = scaler.fit_transform(data[['mpg', 'cyl']])
```

```python
x_minmax
```

**2. Standardization (Z-score)**

Centers data to mean 0, standard deviation 1. Works well even with outliers and doesn't force a bounded range.

Use when

-> SVM & logistic regression

-> PCA & dimensionality reduction

-> Linear models

-> When distribution is roughly Gaussian

-> Most general-purpose ML

Avoid when

Not ideal for heavily skewed data; transform first, then standardize.

```python
from sklearn.preprocessing import StandardScaler
```

```python
data.head()
```

```python
scaler = StandardScaler()
x_standard = scaler.fit_transform(data[['disp', 'hp']])
```

```python
x_standard
```

**3. Robust Scaler**

Uses the median and interquartile range (IQR) instead of mean and std. Highly resistant to outliers.

Use when

-> Data with many outliers

-> Financial / medical datasets

-> Fraud detection

-> Salary / income data

-> When you cannot remove outliers

Avoid when

Not needed if data is already clean and normally distributed.

```python
from sklearn.preprocessing import RobustScaler
```

```python
data.head()
```

```python
scaler = RobustScaler()
x_robust = scaler.fit_transform(data[['drat', 'wt']])
```

```python
x_robust
```

**4. MaxAbs Scaler**

Scales by dividing by the maximum absolute value. Output is in [-1, 1]. Designed for sparse data — does not shift center.

Use when

-> Sparse matrices (text, TF-IDF)

-> Already centered data

-> SVMs on text data

-> NLP feature matrices

Avoid when

Not suited for dense data with skewness or large outliers.

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
data.head()
```

```python
scaler = MaxAbsScaler()
x_maxabs = scaler.fit_transform(data[['qsec', 'vs']])
```

```python
x_maxabs
```

**5. Normalizer(L1/L2)**

Normalizes each sample (row), not each feature (column). Makes each row unit length. Used in text/NLP.

Use when

-> Text classification (TF-IDF vectors)

-> Document similarity

-> Cosine similarity tasks

-> Clustering on sparse data

Avoid when

Do not use when you need feature-level scaling — this works on samples, not features.

```python
from sklearn.preprocessing import Normalizer
```

```python
data.head()
```

```python
normalizer = Normalizer()
x_normalizer = normalizer.fit_transform(data[['am', 'gear', 'carb']])
```

```python
x_normalizer
```
